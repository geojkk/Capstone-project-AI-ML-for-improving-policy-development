```python
import numpy as np
import matplotlib.pylab as plt
```

```python
# Generate synthetic data: heights of 50 students (in cm)
np.random.seed(42)  # for reproducible results
heights = np.random.normal(loc=165, scale=10, size=50)  # average 165 cm, sd=10

# Create a simple line plot
plt.figure(figsize=(10, 5))
plt.plot(heights, marker='o', linestyle='-', color='blue')

# Add labels and title
plt.title("Heights of 50 Students in a Class")
plt.xlabel("Student Number")
plt.ylabel("Height (cm)")

# Display the plot
plt.grid(True)
plt.show()
```
